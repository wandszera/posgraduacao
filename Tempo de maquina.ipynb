# EXECUTAR NO DIRETÓRIO RAIZ DO PROJETO/CÁLCULO

```python
import os
import re

# Função para calcular o tempo de execução a partir de um arquivo OUTCAR
def calcular_tempo_execucao(outcar_path):
    pattern = r"Total CPU time used \(sec\)\s*:\s*([\d\.]+)"

    with open(outcar_path, 'r') as outcar_file:
        outcar_content = outcar_file.read()
        match = re.search(pattern, outcar_content)

    if match:
        return float(match.group(1))
    else:
        return 0.0

# Função para converter segundos para horas
def segundos_para_horas(segundos):
    return segundos / 3600

# Obter o diretório de trabalho atual
pasta_raiz = os.getcwd()

# Lista para armazenar os tempos de execução de cada subdiretório
tempos_execucao = []

# Loop pelos subdiretórios
for subdir, dirs, files in os.walk(pasta_raiz):
    for file in files:
        if file == "OUTCAR":  # Certifique-se de que estamos lidando com o arquivo OUTCAR
            outcar_path = os.path.join(subdir, file)
            tempo_execucao = calcular_tempo_execucao(outcar_path)
            print(f"Subdiretório: {subdir}, Tempo de execução: {tempo_execucao} segundos")
            tempos_execucao.append(tempo_execucao)

# Calcular a soma total e a média dos tempos de execução em horas
if tempos_execucao:
    soma_total_tempos = sum(tempos_execucao)
    media_tempos_execucao = soma_total_tempos / len(tempos_execucao)
    
    # Converter para horas
    soma_total_horas = segundos_para_horas(soma_total_tempos)
    media_horas = segundos_para_horas(media_tempos_execucao)
    
    print(f"Soma total dos tempos de execução: {soma_total_horas:.2f} horas")
    print(f"Média dos tempos de execução: {media_horas:.2f} horas")
else:
    print("Nenhum tempo de execução encontrado nos subdiretórios.")
```
